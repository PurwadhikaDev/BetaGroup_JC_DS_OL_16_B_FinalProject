Analisis akan dilakukan dengan menggunakan 2 pendekatan berikut:

### B. *Predictive Analytics*

Selanjutnya, untuk memprediksi apakah customer akan berhenti/tidak berhenti menggunakan layanan, model machine learning (dalam hal ini supervised learning) akan digunakan. Variable churn ditetapkan sebagai Response Variable/ Target/ Independent Variable, dengan:

        1 : Churn

        0 : Tidak Churn
        
Berdasarkan definisi tersebut, maka target varibale merupakan categorical variable, sehingga model supervised learning yang akan digunakan adalah model klasifikasi. Confusion matrix yang terbentuk sebagai berikut:

| |Predicted Value: 0 (Tidak Churn)|Predicted Value: 1 (Churn)
|--|--|--|
|Actual Value: 0 (Tidak Churn)| True Negative| False Positive (Type 2 Error) |
|Actual Value: 1 (Churn)| False Negative (Type 1 Error) | True Positive |

Type 1 Error (False Positive)
  - Diprediksi churn namun kenyataannya tidak
  - Konsekuensi : kerugian atas biaya untuk menerapkan strategi retensi/preventif

Type 2 Error (False Negative)
  - Diprediksi tidak churn namun kenyataannya churn
  - Konsekuensi : Kehilangan customer, sehingga perusahaan perlu melakukan akusisi customer baru untuk mempertahankan growth rate

Berdasarkan analisis di atas, false negative lebih merugikan daripada false positif karena biaya akusisi > biaya retensi customer, sehingga:
- False negative rate diharapkan mendekati nol
- Negative prediction diharapkan sangat akurat

Oleh karena itu metric yang akan digunakan adalah **recall**

        recall (P) = (True Positive)/(True Positive + False Negative)

Namun, disisi lain type 1 error juga perlu untuk diminimalisir, sehingga biaya untuk menerapkan strategi rentensi tidak terbuang sia-sia. Oleh karenanya, akan digunakan juga metric **F2-score** untuk memilih model dengan recall tertinggi, namun tanpa mengabaikan precision.

       F2-score = (5 * Precision * Recall) / (4 * Precision + Recall)

dimana

       precision = (True Positive)/(True Positive + False Positive)

Dengan demikian, maka **Model akan dipilih berdasarkan nilai recall dan F2-score yang tinggi**

```python
Tenure_q1 = np.percentile(ten_df['Tenure'], 25)
Tenure_q3 = np.percentile(ten_df['Tenure'], 75)
Tenure_iqr = Tenure_q3 - Tenure_q1

Tenure_lower = Tenure_q1 - 1.5*Tenure_iqr
Tenure_upper = Tenure_q3 + 1.5*Tenure_iqr
Tenure_min = ten_df['Tenure'].min()

Tenure_outliers = ten_df[ten_df['Tenure'] > Tenure_upper]
percent_Tenure_outliers = round(len(Tenure_outliers)/len(df)*100, 2)

feature_outliers.append({
    'Feature' : 'Tenure',
    'Minimum value' : Tenure_min,
    'Lower Value' : Tenure_lower,
    'Upper Value' : Tenure_upper,
    'Percentage' : percent_Tenure_outliers
})

pd.DataFrame(feature_outliers)
```

## Churn rate

Churn rate adalah persentase pelanggan yang berhenti menggunakan produk atau layanan suatu perusahaan dalam periode tertentu. Churn rate dapat dipengaruhi oleh berbagai macam faktor. ### by Tenure

Tenure adalah lamanya seseorang menjadi pengguna dari aplikasi e-commerce. Tenure dijelaskan dalam unit bulan. Lamanya tenure seringkali dikaitkan dengan tingkat loyalitas customer terhadap suatu jasa. Terdapat masa - masa krusial dalam waktu seseorang menjadi member suatu aplikasi dimana dia memiliki tendensi untuk churn. Ketika masa - masa krusial ini dapat dipetakan, kita dapat menempatkan strategi retensi secara tepat sebelum churn itu terjadi.

Untuk mempermudah analisa, kami membagi tenure kedalam beberapa kelompok sesuai dengan kuartilnya.

```python
q1 = np.percentile(ten_df['Tenure'], 25)  # 25th percentile
q3 = np.percentile(ten_df['Tenure'], 75)  # 75th percentile

print(f"Q1 (25th percentile): {q1}")
print(f"Median (50th percentile): {ten_df['Tenure'].median()}")
print(f"Q3 (75th percentile): {q3}")
```

Dari keempat kuartil, kami membagi tenure kedalam short tenure, medium-short tenure, medium-long tenure, dan long tenure.

```python
# Define bin edges based on percentiles
bins = [ten_df['Tenure'].min(), q1, ten_df['Tenure'].median(), q3, ten_df['Tenure'].max()]

# Define bin labels
labels = ['Short Tenure (<2 mo)', 'Medium-Short Tenure (3-9 mo)', 'Medium-Long Tenure (10-16 mo)', 'Long Tenure (>16 mo)']

# Apply binning
ten_df['Tenure Category'] = pd.cut(ten_df['Tenure'], bins=bins, labels=labels, include_lowest=True)

# Check distribution
print(ten_df['Tenure Category'].value_counts())
```

Dari bar plot di atas dapat disimpulkan bahwa masa - masa kritis customer dimana churn berpotensi tinggi terjadi adalah pada kurun waktu di bawah 2 bulan. Di atas kurun waktu dua bulan, terlihat loyalitas customer terhadap e-commerce mulai terbentuk secara signifikan dengan adanya penurunan tingkat churn. Oleh karena itu dapat disimpulkan bahwa strategi retensi dapat dipusatkan pada masa - masa kritis selama dua bulan pertama customer dalam e-commerce.
